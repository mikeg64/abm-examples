```python
import matplotlib.pyplot as plt
import numpy as np

price = np.linspace(1, 10, 100)
quantity_demanded = 100 - 10 * price

plt.plot(price, quantity_demanded, label='Demand', color='green')
plt.xlabel('Price')
plt.ylabel('Quantity Demanded')
plt.title('Law of Demand')
plt.grid(True)
plt.legend()
plt.show()
```

```python
price = np.linspace(1, 10, 100)
quantity_supplied = 10 * price - 10

plt.plot(price, quantity_supplied, label='Supply', color='blue')
plt.xlabel('Price')
plt.ylabel('Quantity Supplied')
plt.title('Law of Supply')
plt.grid(True)
plt.legend()
plt.show()
```

```python
price = np.linspace(1, 10, 100)
demand = 100 - 10 * price
supply = 10 * price - 10

plt.plot(price, demand, label='Demand', color='green')
plt.plot(price, supply, label='Supply', color='blue')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.title('Market Equilibrium')
plt.grid(True)
plt.legend()

# Find equilibrium
eq_price = (100 + 10) / 20
eq_quantity = 100 - 10 * eq_price
plt.plot(eq_price, eq_quantity, 'ro')  # equilibrium point
plt.annotate('Equilibrium', xy=(eq_price, eq_quantity), xytext=(eq_price+0.5, eq_quantity+10),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
price = np.linspace(1, 10, 100)
elastic = 100 - 20 * price
unit_elastic = 100 - 10 * price
inelastic = 100 - 5 * price

plt.plot(price, elastic, label='Elastic', linestyle='--')
plt.plot(price, unit_elastic, label='Unit Elastic')
plt.plot(price, inelastic, label='Inelastic', linestyle=':')
plt.xlabel('Price')
plt.ylabel('Quantity Demanded')
plt.title('Price Elasticity of Demand')
plt.grid(True)
plt.legend()
plt.show()
```

```python
quantity = np.linspace(1, 20, 100)
marginal_benefit = 100 / quantity
marginal_cost = 2 * quantity

plt.plot(quantity, marginal_benefit, label='Marginal Benefit', color='green')
plt.plot(quantity, marginal_cost, label='Marginal Cost', color='red')
plt.xlabel('Quantity')
plt.ylabel('Value')
plt.title('Marginal Analysis')
plt.grid(True)
plt.legend()

# Optimal point
optimal_q = np.sqrt(50)
optimal_value = 100 / optimal_q
plt.plot(optimal_q, optimal_value, 'ro')
plt.annotate('Optimal Decision', xy=(optimal_q, optimal_value), xytext=(optimal_q+1, optimal_value+10),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python

from mpl_toolkits.mplot3d import Axes3D

L = np.linspace(1, 100, 50)
K = np.linspace(1, 100, 50)
L, K = np.meshgrid(L, K)
A = 1
alpha = 0.3
beta = 0.7
Q = A * (L ** alpha) * (K ** beta)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(L, K, Q, cmap='viridis')
ax.set_xlabel('Labor')
ax.set_ylabel('Capital')
ax.set_zlabel('Output')
ax.set_title('Cobb-Douglas Production Function')
plt.show()
```
